Plot the histogram, bar chart and pie chart on a sample data set


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Values': [23, 45, 56, 78, 33]
}
df = pd.DataFrame(data)

# 1. Plot a histogram (using random data for demonstration)
random_data = np.random.randn(1000)

plt.figure(figsize=(10, 6))
plt.subplot(1, 3, 1)
plt.hist(random_data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram')

# 2. Plot a bar chart
plt.subplot(1, 3, 2)
plt.bar(df['Category'], df['Values'], color='lightcoral')
plt.title('Bar Chart')

# 3. Plot a pie chart
plt.subplot(1, 3, 3)
plt.pie(df['Values'], labels=df['Category'], autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')

# Display all charts
plt.tight_layout()
plt.show()
```
